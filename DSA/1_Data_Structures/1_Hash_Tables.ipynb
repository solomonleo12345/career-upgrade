**Given an array, find the first recurring number..?**

```python
# Given an array, find the first recurring element in it.
# First approach to solve the problem using brute force method

def first_recurring_element(arr):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                return arr[i]
    return None

# Example usage
array = [2, 5, 1, 2, 3, 5, 1, 2, 4]
result = first_recurring_element(array)
if result:
    print(f"The first recurring element is: {result}")
```

```python
# Given an array, find the first recurring element in it.
# Second approach to solve the problem using a hash set (dictionary) for better efficiency

def first_recurring_element_efficient(arr):
    seen = {}
    for element in arr:
        if element in seen:
            return element
        else:
            seen[element] = True
    return None

# Example usage
array = [2, 5, 1, 2, 3, 5, 1, 2, 4]
result = first_recurring_element_efficient(array)
if result:
    print(f"The first recurring element is: {result}")
```
